# A Look At Kepler's First Law

The three **Kepler's laws** are the fundamental laws of the orbit of the planets around the sun, and they are:

1. The [Law of Orbits](http://hyperphysics.phy-astr.gsu.edu/hbase/kepler.html#c2) - All planets move in elliptical orbits, with the sun at one focus
2. The [Law of Areas](http://hyperphysics.phy-astr.gsu.edu/hbase/kepler.html#c5) - A line that connects a planet to the sun sweeps out equal areas in equal time
3. The [Law of Periods](http://hyperphysics.phy-astr.gsu.edu/hbase/kepler.html#c6) - The square of the period of any planet is proportional to the cube of the semimajor axis of its orbit

## 1. The Law of Orbits

![Keplers First Law](https://www.sciencefacts.net/wp-content/uploads/2022/04/Keplers-First-Law.jpg)

The sun is not in the center of a circular path that passes through a celestial body, but is offset.

The planet follows this eliptical shape, which means that the distance to the sun is always changing.
Thereby the closest point on the elipse to the sun is also called **Aphel** or **Aphelion**, whereas the farthest point on the ellipse to the sun is also called **Perihel** or **Perihelion**.

The deviation of an ellipse from the circular form is also called **eccentricity** $\epsilon$ and results from the equation:

\begin{align}
    \epsilon\equals\frac{e}{a}
\end{align}

The eccentricity is always between **0** and **1**, where 0 corresponds to the circular form and a correspondingly higher value represents a greater deviation from the circular form.

With the earth orbit the eccentricity $\epsilon$ is for example **0.0167**, which means that the sun lies around **1.6%** of the large half axis laterally of the center.
Thus, the Earth's orbit is "almost" a perfect circle, with only the Sun lying slightly to the side of the center $M$.
Some more examples of different eccentricity values for different planets orbiting the Sun are:

|----------------|--------------|

## 2. The Law of Areas

![Keplers Second Law](https://www.sciencefacts.net/wp-content/uploads/2022/04/Keplers-Second-Law.jpg)

In its orbit around the Sun, the speed of a planet varies.
Near the perihelion, the planet has a higher speed than near the aphelion.
Nevertheless, the area enclosed by the travel ray from the planet to the sun is the same over the same time.

Not very scientifically and rather simplified illustrated this means that the enclosed area (triangle) near the perihelion is short but wide, while near the aphelion it is long but thin.
Thus, the equality can be equated in a simplified way.

Since the elliptical orbit of the earth around the sun is quite close to a perfect circle (small eccentricity), the velocity difference between perihelion (30.29 km/s) and aphelion (29.29 km/s) is very small.

## 3. The Law of Periods

![Keplers Third Law](https://www.sciencefacts.net/wp-content/uploads/2022/04/Keplers-Third-Law.jpg)

The orbital periods to the square of the planets behave to each other like the mean distances to the sun in third power.

\begin{align}
    \frac{T^2_1}{T^2_2}\equals\frac{a^3_1}{a^3_2}
\end{align}

Or also: The ratio of the orbital period (in second power) to the mean distance to the sun (in third power) is always the same, for all planets in the solar system.

\begin{align}
    \frac{T^2_1}{a^3_1}\equals\frac{T^2_2}{a^3_2}\equals{C}
\end{align}

This ratio is also called the Kepler constant and is different for each central star.
Thereby applies to the central star sun:

\begin{align}
    {C_{sun}}\equals{2.97*{10{^{-19}}}}\frac{s^2}{m^3}
\end{align}

for the central star Jupiter applies:

\begin{align}
    {C_{sun}}\equals{3.1*{10{^{-16}}}}\frac{s^2}{m^3}
\end{align}

and for the central star Earth applies:

\begin{align}
    {C_{sun}}\equals{9.91*{10{^{-14}}}}\frac{s^2}{m^3}
\end{align}

Kepler's laws assume that the mass of the central body is significantly greater than the mass of the orbiting bodies.
The elliptical orbit is forced by the gravitational force, which causes an acceleration that brings a mass body in the vicinity of another heavy body onto the aforementioned orbit.
